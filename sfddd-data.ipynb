# State Farm Distrated Driver Detection

## Create training data and save files

```python
#common 
import pandas as pd
import os
import numpy as np

imgList = '/usr/local/hadoop/project/sfddd/driver_imgs_list.csv'
trainFile1 = '/usr/local/hadoop/project/sfddd/trainset1.csv'
trainFile2 = '/usr/local/hadoop/project/sfddd/trainset2.csv'
trainFile3 = '/usr/local/hadoop/project/sfddd/trainset3.csv'
imgSource = '/usr/local/hadoop/project/sfddd/imgs.zip'
train1ImgsFolder = '/usr/local/hadoop/project/sfddd/train1Imgs'
train2ImgsFolder = '/usr/local/hadoop/project/sfddd/train2Imgs'
train3ImgsFolder = '/usr/local/hadoop/project/sfddd/train3Imgs'
train1ImgsList = '/usr/local/hadoop/project/sfddd/train1Imgs/lst.txt'
```

### Break up training data into multiple lists (trainFileX csv)

```python

allImg = pd.read_csv(imgList)

def getTrainSet(num,rs):
    return allImg.sample(n=num,random_state=rs)
def saveTrainSet(s,f):
    try:
        os.remove(f)
    except OSError:
        pass

    with open(f,'w') as target:
        target.write("\n".join(('train/' + s['classname']  + "/"+ s['img']).tolist()))

set1 = getTrainSet(500,2000)
set2 = getTrainSet(500,2001)
set3 = getTrainSet(500,2002)

saveTrainSet(set1,trainFile1)
saveTrainSet(set2,trainFile2)
saveTrainSet(set3,trainFile3)
```

### Convert images in train set1 to greyscale

```python
import zipfile
from PIL import Image # $ pip install pillow
from matplotlib.pyplot import imshow
import numpy as np
%matplotlib inline

save_list =[]
with zipfile.ZipFile(imgSource, 'r') as archive:
    lines =[]
    with open(trainFile1) as f:
        lines = f.read().splitlines()
        for x in lines:
            with archive.open(x) as file:
                split =  x.split('/')
                save_to = os.path.join(train1ImgsFolder, split[len(split)-1])
                img = Image.open(file)
                gry = img.convert('L')
                #imshow(np.asarray(gry),cmap='Greys_r')
                gry.save(save_to)
                save_list.append(save_to)
                
with open (train1ImgsList, 'w+') as f:
    for item in save_list:
        f.write("%s\n" % item)
print "Wrote " + str(len(save_list)) + " lines to " + train1ImgsList
    

#import matplotlib.pyplot as plt
#plt.imshow(i, cmap=plt.cm.gray)
```
